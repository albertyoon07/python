```python
from selenium import webdriver
# from urllib.request import HTTPError
# import requests  
from bs4 import BeautifulSoup 
import pandas as pd
import time
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
pd.set_option('max_rows', 10000)
pd.set_option('max_columns', 10000)
```

```python
# final code
driver = webdriver.Chrome('C:/Users/seob6/Desktop/chromedriver/chromedriver_win32/chromedriver.exe')
results = []
url = 'https://casenote.kr/search/?q=%EC%84%B8%EC%9B%94%ED%98%B8&sort=1&cc=0&ct=0&pr=0&pf=&pt=&p={}'
for i in range(1, 20):
    link = url.format(i)
    driver.get(link)  # 해당 링크 open
    driver.implicitly_wait(5)
    
    html = driver.page_source  # 현재 웹페이지 소스 코드
    soup = BeautifulSoup(html, 'lxml')   # 소스 코드 정적 수집
    for i in range(0, 10):
        try:
            title = soup.select('div.searched_item a')[i].text
            link = soup.select('#mainbar > div > div > a')[i]['href']
            results.append([title, link])
            driver.implicitly_wait(1.5)
        except:
            driver.implicitly_wait(5)
            break
driver.quit()
df = pd.DataFrame(data=results, columns=['title', 'link'])
df
```

```python
contents = []

for link in df['link']:
    driver = webdriver.Chrome('C:/Users/seob6/Desktop/chromedriver/chromedriver_win32/chromedriver.exe')
    url = 'https://casenote.kr'+link
    driver.get(url)
#     time.sleep(0.5)
    driver.implicitly_wait(5)
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    content = soup.find(id='mainbar').get_text()
    contents.append(content)
    driver.close()
    
df['content'] = contents
df
```
